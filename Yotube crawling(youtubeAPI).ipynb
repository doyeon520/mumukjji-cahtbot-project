# youtube-API이용(1) 

```python
import os
from apiclient.discovery import build
import pprint
```

```python
YOUTUBE_API_KEY = '자신의 yotubeAPI' 

youtube = build('youtube', 'v3', developerKey=YOUTUBE_API_KEY)
```

```python
# 키워드 매개변수로 매개변수를 지정하고
# search.list 메서드를 호출합니다.
# list() 메서드를 실행하면 googleapiclient.http.HttpRequest가 반환됩니다. 
# execute() 메서드를 실행하면 실제 HTTP 요청이 보내지며, API 응답이 반환됩니다.
search_response = youtube.search().list(
    part = 'snippet',
    q = '파이썬', #검색어
    type = 'video',
    maxResults = 10 #개수
).execute()
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(search_response)
```

```python
# search_response는 API 응답을 JSON으로 나타낸 dict 객체입니다.
for item in search_response['items']:
    # 동영상 제목을 출력합니다.
    print(item['snippet']['title'],item['snippet']['channelTitle'])
```

# youtube-API이용(2) youtube 본문전체, 태그 크롤링

```python
response = youtube.videos().list(part='snippet', id='8B8boXS8_iU').execute()
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(response)
```

```python
for item in response['items']:
    # 동영상 제목을 출력합니다.
    print(item['snippet']['description'])
    print(item['snippet']['tags'])
```
