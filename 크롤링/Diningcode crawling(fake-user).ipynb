```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import urllib
import requests 
import json
import re
import os
import numpy as np
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from fake_useragent import UserAgent
from selenium.webdriver.chrome.options import Options
```

```python
word = ['관자찜 맛집']
#headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/28.0.1467.0 Safari/537.36'} 
서칭 = []
links = []
iword = []
for i in word:
    print(i)
    word_encode = urllib.parse.quote(i)
    
    options = webdriver.ChromeOptions()   
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko")
    driver = webdriver.Chrome('chromedriver', chrome_options=options)
    driver.get('https://www.diningcode.com/list.php?query={}'.format(word_encode)) 

    time.sleep(2)
    
    rlink = []
    
    while True:
        try:
            더보기 = driver.find_element_by_xpath('//*[@id="div_list_more"]/span/span')
            더보기.click()
            time.sleep(2)
        except:
            break
            
    while True:
        try:
            m = driver.find_element_by_css_selector('#div_list_more')
            m.click()
            time.sleep(1)
        except:
            break
            
    #print(word_encode)
    html = driver.page_source
    #print(html)
    
    #response = requests.get(rrr, headers)
    soup = BeautifulSoup(html, 'html.parser')
    for link in soup.find_all('a'):
        if 'profile' in link.get('href'):
            rlink.append('https://www.diningcode.com/' + link.get('href'))
            iword.append(i)
    print(len(rlink))
    links += rlink
print(links)
```

```python
np.savetxt('dining_code_url.txt', links,fmt='%s')
```

```python
test = pd.DataFrame(columns=['서칭','맛집', '평점','주소','번호','영업시간','메뉴정보','리뷰'])
test
```

```python
for final in range(len(result_list)):
    #driver=webdriver.Chrome('chromedriver.exe')
    opts = Options()
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko")
    driver = webdriver.Chrome(chrome_options=opts)
    driver.get(result_list[final])
    #html_doc = driver.page_source
    #soup = BeautifulSoup(html_doc, 'html.parser')
    #맛집 이름  
    try:    
        title = driver.find_element_by_xpath('//*[@id="div_profile"]/div[1]/div[2]/p').text
    except:
        title = None
        
    #맛집 평점
    try:
        count = driver.find_element_by_xpath('//*[@id="lbl_review_point"]').text.replace('\n', '')
    except:
        count = None
    # info = soup.find("tbody").get_text().replace('\n', '')
    # print(info)
    # 맛집 주소
    
    try:
        address = driver.find_element_by_xpath('//*[@id="div_profile"]/div[2]/ul/li[1]').text
    except:
        address = None
        
    # 전화 번호
    try:
        number = driver.find_element_by_xpath('//*[@id="div_profile"]/div[2]/ul/li[2]').text
    except:
        number = None
    # 영업 시간
    try:
        time = driver.find_element_by_xpath('//*[@id="div_detail"]/div[1]').text.replace('\n', '')
    except:
        time = None
    # 메뉴 정보
    try:
        info = driver.find_element_by_xpath('//*[@id="div_detail"]/div[2]').text
    except:
        info = None
    
    try:
        menu = info.replace('메뉴정보','').replace('더보기', '').replace('\n', '')
    except:
        menu = None
    # 리뷰 
    
    try:
        review = driver.find_element_by_xpath('//*[@id="div_profile"]/div[8]').text.replace('\n', '')
    except:
        review = None
        
    
    test.loc[final] = [iword[final],title,count,address,number,time,menu,review]
    
    driver.close()
```
