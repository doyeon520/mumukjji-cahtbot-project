```python
#가져온 데이터 데이터프레임으로 만들기
import pandas as pd
def get_tablebylocation(place):
    df = pd.DataFrame(columns=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r'])
    gu = place.split()[0]
    dong = place.split()[1]
    conn = pymysql.connect(host='127.0.0.1', user='root', db='ourmz', charset='utf8',password='')
    # Connection 으로부터 Cursor 생성
    curs = conn.cursor()
    sql = "select * from ourmz.store"
    curs.execute(sql)
    rows = curs.fetchall()
    #print(rows)
    for i in range(len(rows)):
        a = rows[i]
        df.loc[i] = a
    df=df.loc[df['m']==gu]
    df = df.loc[df['n'].str.contains(dong)]
    df = df.reset_index(drop=True)
    return df
```
