```python
#자막 전처리
import os
#from apiclient.discovery import build
import pprint
from pytube import YouTube

import warnings

warnings.filterwarnings(action='error')
caption_list = []
index = 0
yt = YouTube('https://www.youtube.com/watch?v=kK4I76sBajA')
caption = yt.captions.get_by_language_code('ko')
for line in str(caption.generate_srt_captions()).split('\n'):
    if index == 0:
        caption_list.append({})
    if index in (1, 2):
        caption_list[len(caption_list)-1][('time', 'caption')[index-1]] = line
    index += 1
    if line == '':
        index = 0
b=[]
for dic in caption_list:
    #print('{}'.format(dic['caption']))
    a=dic['caption']
    b.append(a)
    #print(a)    
print(b)
```
